## k Mean Clustring

in k mean clustring unlabel data is label by mechine learning modle into label data into catagory
E.g :  age devide into different catagory after doing clustring\
5 to 15 child\
17 to 30 young\
30 to 50 old


Oerall data of age is ublaabel data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as  plt

df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
# modle

from sklearn.cluster import KMeans
# sklearn sa ham na KMewan algrothaim impott kia ha

#normalize karny ka lia ham na ss ko import kia ha
from sklearn.preprocessing import StandardScaler

# import the data in feature
X = df[['age', 'fare']]

# Remove the NaN from x
X = pd.get_dummies(X, columns=['age','fare'])
#Scaled the data 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform Clustring
modle = KMeans(n_clusters=5)
modle.fit(X_scaled)

# Asssign cluster label to data
df['cluster']= modle.labels_
```

```python
X.head()
```

```python
df['cluster'].unique()
```

```python
plt.scatter(df['age'],df['fare'],c=df['cluster'],cmap='rainbow')
plt.scatter(modle.cluster_centers_[:, 0],modle.cluster_centers_[:, 1], color='black')
plt.xlabel("age")
plt.ylabel("fare")
plt.show
```
