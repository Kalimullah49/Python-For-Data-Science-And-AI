# 03_How to rename columns


```python
df = pd.DataFrame({'A col': [1,2,3,8,9,0,22],'B col':[4,5,6,7,8,9,10]})
df.head()
```

```python
df.rename(columns={'A col':'col_a','B col':'col_b'},inplace=True)
df
```

```python
 #2nd way to rename colum
df.columns={'col_aa','col_bb'}
df
```

```python
# to replace any character string
df.columns=df.columns.str.replace('_',' ') #underscore ko space sa change kia ha
df
```

```python
df.columns=df.columns.str.replace(' ','_') #yahan pa again space change ho ge underscore sa
df
```

```python
#Addinf prefix to columns
df.add_prefix('baba_')
```

```python
#Addinf sufix to columns
df.add_suffix('_baba')
```

# 8-slect a colum by data type
