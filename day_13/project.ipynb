```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Pairplot
sns.pairplot(df)
plt.show()
```

```python
# 2. Heatmap (Correlation Matrix)
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Your data
launch_angles = [38.73, 67.54, 56.60, 49.93, 27.80, 27.80, 22.90, 63.31, 50.06, 55.40]
horizontal_ranges = [36.15, 37.23, 24.88, 40.90, 66.73, 18.91, 24.26, 51.64, 21.34, 12.70]

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(launch_angles, horizontal_ranges, color='blue', marker='o')
plt.title('Launch Angle vs Horizontal Range')
plt.xlabel('Launch Angle (degrees)')
plt.ylabel('Horizontal Range (meters)')
plt.grid(True)
plt.show()
```
