```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
df=sns.load_dataset("titanic")
```

```python
x = df[['pclass','sex','sibsp','parch','fare']]
y=df['survived']

# x ke base pa y predict krna ha
```

```python
x = pd.get_dummies(x,columns=['sex'])
# x ko ham na dummies v ma convert kar lia (bonary variable ) x=1 and y =0 (integer value)
#Machine learning ke pre processing mein categoryableb variable ko integer mein kyon convert Kiya jata hai

# Ans ;computer ko integer zaya pasand han 
# e.g passenger no 12 insted of her name
```

```python
x.isnull().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.8)
```

```python
modle = DecisionTreeClassifier()
modle.fit(x_train,y_train)
```

```python
y_pred = modle.predict(x_test)
```

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
```

## k nearest Neighbours (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier

modle = KNeighborsClassifier(n_neighbors=3)

modle.fit(x_train,y_train)

y_pred = modle.predict(x_test)

cm = confusion_matrix(y_test,y_pred)
cm

```

###### **from sklearn.neighbors import KNeighborsClassifier:** Yeh line code mein se KNeighborsClassifier class ko import karta hai, jo KNN algorithm ko implement karta hai.

###### **model = KNeighborsClassifier(n_neighbors=3)**: Ek naya KNeighborsClassifier object banaya jata hai, jise 'model' variable mein store kiya gaya hai. n_neighbors parameter se hamne decide kiya hai ke kisi point ke classification mein kitne neighbors ko consider karna hai, yahan par 3 neighbors liye gaye hain.

###### **model.fit(x_train, y_train):** Training data (x_train) aur uski corresponding labels (y_train) ka istemaal karke model ko train kiya jata hai.

###### **from sklearn.metrics import confusion_matrix:** Yeh line confusion_matrix function ko import karta hai, jo model ki predictions aur actual labels ke beech ki confusion matrix generate karta hai.

###### **cm = confusion_matrix(y_test, y_pred):** Confusion matrix ko actual test labels (y_test) aur predicted labels (y_pred) ke beech generate kiya jata hai, aur 'cm' variable mein store kiya jata hai.

###### **cm:** Yeh line code confusion matrix ko print karta hai, jo model ki performance ko evaluate karne mein madad karta hai.##

```python
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Generate some example data (replace this with your actual data loading process)
# Assuming x_train is your feature matrix and y_train is your target variable
x_train = np.random.rand(100, 10)  # Example feature matrix with 100 samples and 10 features
y_train = np.random.randint(0, 2, 100)  # Example target variable with binary labels

# Define the models to be evaluated
models = [LogisticRegression(), KNeighborsClassifier(), RandomForestClassifier(), SVC()]
names = ["Logistic Regression", "KNeighbors Classifier", "Random Forest", "SVC"]

# Perform k-fold cross-validation for each model
k = 10

for name, model in zip(names, models):
    cv_scores = cross_val_score(model, x_train, y_train, cv=k)
    print(f"{name} CV accuracy: {np.mean(cv_scores):.3f} +/- {np.std(cv_scores):.3f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define your parameter grid
param = {'C': [0.1, 1, 10],
         'kernel': ['linear', 'rbf'],
         'gamma': [0.1, 1, 10]}

# Create your model
model = SVC()

# Perform GridSearchCV
grid_search = GridSearchCV(model, param, cv=5)
grid_search.fit(x_train, y_train)

# Print best parameters and best cross-validation score
print("Best parameters: {}".format(grid_search.best_params_))
print("Best cross-validation score: {:.3f}".format(grid_search.best_score_))
```
